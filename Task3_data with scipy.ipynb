```python
import pandas as pd
import numpy as np
dd=pd.read_csv('result_withoutTotal.csv',index_col=0)
dd.head()
```

```python
dd.head()
iris1=pd.DataFrame(dd)
iris1['Total']=0.05*(iris1.Ass1+iris1.Ass3)+0.15*(iris1.Ass2+iris1.Ass4)+0.6*iris1.Exam #finding total
iris1.loc[iris1.Total>100,['Total']]=100           #if total is above 100 it is set to be 100
iris1.head()
```

```python
iris1['Final']=iris1.Total.round().astype(int) #finding final by rounding the figure to nearest integer
iris1.head()
```

```python
iris1.loc[(iris1.Final>44) & (iris1.Exam<48),['Final']]=44 #If a student failed the hurdle (Exam >= 48), the max Final is 44. 
                                                            #No change to Final scoreif Final <44
iris1.head()
```

```python
iris1.loc[(iris1.Final>79.45) & (iris1.Final<=100),['GradeColoumn']]='HD'
iris1.loc[(iris1.Final>69.45) & (iris1.Final<=79.45),['GradeColoumn']]='D'
iris1.loc[(iris1.Final>59.45) & (iris1.Final<=69.45),['GradeColoumn']]='C'
iris1.loc[(iris1.Final>49.45) & (iris1.Final<=59.45),['GradeColoumn']]='P'
iris1.loc[(iris1.Final>0) & (iris1.Final<=49.45),['GradeColoumn']]='N'
iris1.head()

```

```python
# saving the dataframe 
iris1.to_csv('results_updated.txt')
```

```python
#collecting details of students with exam<48.
iris2=iris1.loc[(iris1['Exam']<48),['Ass1','Ass2','Ass3','Ass4','Exam','Total','Final','GradeColoumn']] 
iris2.head()
```
